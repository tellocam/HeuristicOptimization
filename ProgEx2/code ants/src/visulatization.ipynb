```julia
using Plots
include("../src/ACS_fun.jl")
include("../bin/ACS.jl")
```

```julia
function spy_plot(matrix, threshold=0.0)
    spy_matrix = abs.(matrix) .> threshold
    indices = findall(spy_matrix)

    x = [index[2] for index in indices]
    y = [index[1] for index in indices]

    scatter(x, y, ratio=1, legend=false, xlabel="Column Index", ylabel="Row Index", title="Spy Plot", markersize=0.2)
    
end
```

```julia
folder_path = "../data/datasets/inst_tuning/"
file_names = readdir(folder_path)

for file_name in file_names
    if endswith(file_name, ".txt")
        file_path = joinpath(folder_path, file_name)
        matrix = readSPSolutionFile(file_path).A0
        spy_plot(matrix)
        savefig(joinpath(folder_path, "spy_plot_$(file_name).png"))  # Save each plot as a PNG file
    end
end
```

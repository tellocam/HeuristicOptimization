```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pylab import rcParams
```

```python
rcParams['figure.figsize'] = 7, 4
plt.rc('text', usetex=True)
plt.rcParams['text.usetex']=True
plt.rc('font', family='sans-serif', weight='bold')

SMALL_SIZE = 6
MEDIUM_SIZE = 7
BIGGER_SIZE = 8

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)    # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
plt.rcParams['figure.dpi'] = 500
```

```python
# Step 1: Read the CSV File
df5 = pd.read_csv('inst_test/heur005_n_160_m_4015.txt', header=None)
df35 = pd.read_csv('inst_test/heur035_n_345_m_17798.txt', header=None)
df32 = pd.read_csv('inst_test/heur032_n_330_m_9770.txt', header=None)

# Step 2: Convert Binary Strings to NumPy Array
data5 = df5.apply(lambda x: list(map(int, list(x[0]))), axis=1).to_list()
data35 = df35.apply(lambda x: list(map(int, list(x[0]))), axis=1).to_list()
data32 = df32.apply(lambda x: list(map(int, list(x[0]))), axis=1).to_list()
```

```python
plt.subplot(1, 3, 1)
plt.spy(data5, markersize=0.1)
plt.title('Test Instance heur005')
plt.legend(["n=160, m=4015"], loc="lower left")
plt.ylabel('Row Index')

plt.subplot(1, 3, 2)
plt.spy(data35, markersize=0.1)
plt.title('Test Instance heur035')
plt.legend(["n=345, m=17798"], loc="lower left")
plt.xlabel('Column Index')

plt.subplot(1, 3, 3)
plt.spy(data32, markersize=0.05)
plt.title('Test Instance heur032')
plt.legend(["n=330, m=9770"], loc="lower left")
plt.tight_layout()

plt.savefig("../../latex1/figures/spy_adjacency.pdf", format='pdf')
```

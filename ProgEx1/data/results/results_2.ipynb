```python
# Assuming 'x_column' and 'y_column' are the columns you want to plot
x_column = 'm'
y_column = 'obj_val'

unique_ns = df['algo'].unique()
# cmap = plt.cm.get_cmap('Blues')
# n_colors = {unique_n: cmap(i / (len(unique_ns) - 1)) for i, unique_n in enumerate(unique_ns)}

print("Unique 'algo' values:", unique_ns)
# print("n_colors dictionary:", n_colors)
colors_dict = {
    'grasp': 'Red',
    'gvns': 'Blue',
    'vnd': 'green'
}

plt.grid(alpha=0.3)
for n, n_df in df.groupby('algo'):
    x = n_df[x_column]
    y = n_df[y_column]
    color = colors_dict[n]
    
    plt.scatter(n_df[x_column], n_df[y_column], label=f'{n}', marker='o', color=color, s=20, alpha=0.2)

plt.xlabel("Number of Edges m / -")
plt.ylabel("Objective Value / -")
plt.title("Instances Solved with GRASP, GVNS and VND Algorithm")
if save_figures_to_latex == True :
    plt.savefig("../../latex1/figures/vnd_grasp_gvns_edges_vs_obj.pdf", format='pdf')
plt.legend()
```
